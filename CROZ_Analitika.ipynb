```python
#Importing libraries

import pandas as pd
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot') 
print('Matplotlib version: ', mpl.__version__) 
```

### Creating Dataframes from Excel Data

```python
all = pd.DataFrame({'Month' : ['JAN', 'FEB', 'MAR', 'APR','MAY', 'JUN', 'JUL', 'AUG', 'SEP'],
                    'Total_all' : [4406, 3511, 1306, 5461, 3630, 76, 1013, 3628, 6297],
                    'Food & Drink' : [2029, 3480, 1270, 2845, 2224, 0, 0, 3078, 4606],
                    'Luxuries' : [2130, 0, 36, 2418, 1217, 76, 1013, 550, 1559],
                    'Cigarettes' : [246, 180, 0, 198, 189, 0, 0, 0, 132]                  })
all = all.set_index('Month')
all
```

```python
fd = pd.DataFrame({'Month' : ['JAN', 'FEB', 'MAR', 'APR','MAY', 'JUN', 'JUL', 'AUG', 'SEP'],
                    'Total_fd' : [2029, 3480, 1270, 2845, 2224, 0, 0, 3078, 4606],
                    'HEM' : [1465, 2156, 1270, 1919, 1286, 0, 0, 2173, 3779],
                    'ICA' : [282, 662, 0, 462, 468, 0, 0, 452, 413]    })
fd = fd.set_index('Month')
fd    
```

```python
cig = pd.DataFrame({'Month' : ['JAN', 'FEB', 'MAR', 'APR','MAY', 'JUN', 'JUL', 'AUG', 'SEP'],
                    'Total_cig' : [246, 180, 0, 198, 189, 0, 0, 0, 132],                
                    'Lucky' : [114, 114, 0, 0, 57, 0, 0, 0, 0],
                    'Marlboro' : [132, 66, 0, 198, 132, 0, 0, 0, 132]
                     })
cig = cig.set_index('Month')
cig   
```

```python
lux = pd.DataFrame({'Month' : ['JAN', 'FEB', 'MAR', 'APR','MAY', 'JUN', 'JUL', 'AUG', 'SEP'],
                    'Total_lux' : [2130, 0, 36, 2418, 1217, 76, 1013, 550, 1559],                
                    })
lux = lux.set_index('Month')
lux   
```

### Visualizing the Data

```python
colors_list = ['silver', 'teal','skyblue']


all['Total_all'].plot(kind='pie',
                            figsize=(7, 7),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=False,       
                            colors=colors_list) 
                         
                            


plt.title('Total spent per month', y=1.1) 

plt.axis('equal') 

plt.show()
```

```python
colors_list = ['silver','teal','lightskyblue', 'lightgreen']

all.plot(kind='area',
         stacked=False,
         figsize=(10, 6), 
         colors=colors_list)

plt.title('Monthly Total per Category')
plt.ylabel('Total')
plt.xlabel('Month')

plt.show()
```

```python
cig.plot(kind='bar', 
         figsize=(10, 6), 
         color=['silver','orange','red'])

plt.xlabel('Month') 
plt.ylabel('Total')
plt.title('Monthly spend per category', y=1.1) 

plt.show()
```

```python
fd.plot(kind='bar', 
         figsize=(10, 6), 
         color=['silver','teal','lightgreen'])

plt.xlabel('Month') 
plt.ylabel('Total')
plt.title('Monthly spend per category', y=1.1) 

plt.show()
```

```python
all.plot(kind='bar', 
         figsize=(10, 6), 
         color=['silver','teal','lightblue', 'lightgreen'])

plt.xlabel('Month') 
plt.ylabel('Total')
plt.title('Monthly spend per category', y=1.1) 

plt.show()
```

```python
lux.plot(kind='bar', 
         figsize=(10, 6), 
         color=['lightblue'])

plt.xlabel('Month') 
plt.ylabel('Total')
plt.title('Monthly spend on luxury', y=1.1) 

plt.show()
```
